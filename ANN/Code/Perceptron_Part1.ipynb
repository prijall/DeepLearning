```python
#@ Visualizing the data

sns.scatterplot(data=df, x='cgpa', y='resume_score', hue='placed')
```

```python
X=df.iloc[:, 0:2]
y=df.iloc[: ,-1]
```

```python
#@ Using Sklearn library to train perceptron:

from sklearn.linear_model import Perceptron

p= Perceptron()
```

```python
p.fit(X, y)
```

```python
p.coef_
```

- where 'coef_' helps to find the weights in our perceptron. Here, '40.26' is weight for 'cgpa' and '-36' is weight for iq  

```python
p.intercept_
```

- where 'intercept_' gives bias of the perceptron. Here the value od b is -25.

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(X.values, y.values, clf=p, legend=2)
```
