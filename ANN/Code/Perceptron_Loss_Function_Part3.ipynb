```python
from sklearn.datasets import make_classification
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

X, y= make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_classes=2, n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=15)
```

```python
#@ Visualizing the dataset:

plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', s=100)
```

```python
#@ Finding the values of weights and bias in such a way that the loss function of a perceptron is minimum.

def Perceptron(X, y):

    w1=w2=b=1 #Assuming the initial weights and bias to be 1.
    learning_rate=0.1  #for implementation of gradient descent.

    for j in range(1000):
        for i in range(X.shape[0]):

            #checking the condition
            z=w1*X[i][0] + w2*X[i][1] + b 

            if z*y[i]<0:
                #upadating the values of w1 and w2:
                w1=w1+learning_rate*y[i]*X[i][0]
                w2=w2+learning_rate*y[i]*X[i][1]
                b=b+learning_rate*y[i]

    return w1, w2, b
```

```python
w1, w2, b=Perceptron(X, y)
```

```python
m = -(w1/w2)
c = -(b/w2)

print(m,c)
```

```python
x_input=np.linspace(-3, 3, 100)
y_input=m*x_input +c 

plt.figure(figsize=(10, 6))
plt.plot(x_input, y_input, color='red', linewidth=3)
plt.scatter(X[:, 0], X[0:, 1], c=y, cmap='winter', s=100)
plt.ylim(-3, 2)
```
