```python
import cv2 #to read image from disk
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread('MyImage.jpg')
```

```python
# Cropping the img
img = img[50:250,40:240, :]

#converting image into greyscale:
img_grey=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_grey, cmap='gray')
```

```python
# Converting image into 25x25
img_grey_small=cv2.resize(img_grey, (25,25))
plt.imshow(img_grey_small, cmap='gray')
```

```python
print(img_grey_small)
```
